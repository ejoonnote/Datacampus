```python
def get_outlier(df, column, weight):
    #정상 거래에 대해서만 outlier 제거
    fraud = df[df['Class'] == 0][column]
    #25분위와 75분위 계산
    q25 = fraud.quantile(0.25)
    q75 = fraud.quantile(0.75)
    iqr = q75 - q25
    lowest= q25-iqr*weight
    highest = q75 + iqr*weight
    
    return fraud[(fraud > highest) | (fraud < lowest)].index
```

```python
# 데이터 전처리 함수
def get_preprocessed_df(df):
    #1. Time 컬럼 삭제
    df2 = df.copy()
    df2 = df2.drop('Time',axis=1)
    #2. Amount 칼럼을 scaling
#     #2-1. StandardScaler활용
#     stans = StandardScaler()
#     df2.Anmount = stans.fit_transform(df2.Amount.values.reshape(-1,1))
    
    #2-2 log-scale로 변경
    df2['Amount'] = np.log1p(df.Amount)
    
    # V17 outlier 제거
    out_idx = get_outlier(df2, 'V17',1)
    df2.drop(out_idx,axis=0, inplace=True)
    return df2
```

```python
get_preprocessed_df(df)
```

```python
df2 = get_preprocessed_df(df)
```

```python
from sklearn.model_selection import train_test_split
df_x=df2.drop('Class',axis=1)
df_y = df2.Class

X_train, X_test, y_train, y_test = train_test_split(df_x, df_y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score

def get_clf_eval(y_test , pred):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test, pred)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```
